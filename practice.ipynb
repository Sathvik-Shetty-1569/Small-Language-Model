```python
from torch.utils.data import Dataset , DataLoader

class SimpleDataset(Dataset):
    def __init__(self, maxlength ,stride,text,tokenizer):
        self.input_id = []
        self.target_id = []

        token_id = tokenizer.encode(text,allowed_special={'<|endoftext|>'})

        for i in range(0,len(token_id)-maxlength,stride):
            input_chunk = token_id[i:i+maxlength]
            target_chunk = token_id[i+1:i+maxlength+1]
            self.input_id.append(torch.tensor(input_chunk))
            self.target_id.append(torch.tensor(target_chunk))

    def __len__(self):
        return len(self.input_id)
    
    def __getitem__(self, index):
        return self.input_id[index],self.target_id[index]
    

    
def create_dataload(text,batch_size = 4,stride = 128,shuffle = True, maxlength = 256,drop_last =True ,numberofworker = 0):

    tokenizer = tiktoken.get_encoding('gpt2')

    dataset = SimpleDataset(maxlength,stride,text,tokenizer)

    dataloader = DataLoader(dataset,
                            batch_size=batch_size,
                            shuffle=shuffle,
                            drop_last=drop_last,
                            num_workers=numberofworker)
    
    return dataloader
```

```python
with open('the-verdict.txt',encoding='utf-8') as f:
    rawdata = f.read()

import torch
import tiktoken
from torch.utils.data import Dataset,DataLoader

class Simpledataset(Dataset):
    def __init__(self,text,tokenizer,maxlength,stride):
        self.input_id = []
        self.target_id = []

        token_ids = tokenizer.encode(text,allowed_special={'<|endoftext|>'})
        
        for i in range(0,len(token_ids)-maxlength,stride):
            input_chunks = token_ids[i:i+maxlength]
            target_chunks = token_ids[i+1:i+maxlength+1]
            self.input_id.append(torch.tensor(input_chunks))
            self.target_id.append(torch.tensor(target_chunks))

    def __len__(self):
        return len(self.input_id)
    
    def __getitem__(self, index):
        return self.input_id[index],self.target_id[index]
    

def create_dataloader(text,batch_size = 4,stride = 4,maxlength = 4,shuffle=False,drop_last=True,number_of_workers=0):

        tokenizer = tiktoken.get_encoding('gpt2')

        dataset = Simpledataset(text,tokenizer=tokenizer,maxlength=maxlength,stride=stride)

        dataloader = DataLoader(
            dataset=dataset,
            batch_size=batch_size,
            shuffle=shuffle,
            drop_last=drop_last,
            num_workers=number_of_workers
        )

        return dataloader
    
dataloader = create_dataloader(text=rawdata)
data_iter = iter(dataloader)
inputs,targets = next(data_iter)
print(inputs)
print(targets)
```
